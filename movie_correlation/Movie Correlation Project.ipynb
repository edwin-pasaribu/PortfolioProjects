```python
# Import Libraries
import numpy as np

import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjust the configuration of the plots created

# read in the data
df = pd.read_csv(r'D:\Edwin\Porto\3. Movie Dataset - Python\movies.csv')
```

```python
# lets look at the data
df.head()
```

```python
# checking for missing data (if any)
print(len(df))
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# dropping the NA/missing values and recheck
df = df.dropna()
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
print(len(df), 'rows')
```

```python
# checking the datatypes
df.dtypes
```

```python
# change the data types from float64 to int64

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# correcting the year column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
# chnge the option in pandas to show all rows instead, and sort the df by gross descending
pd.set_option('display.max_rows', None)
df.sort_values(by=['gross'], inplace=True, ascending=False)
df.head()
```

```python
# drop any duplicates if any
df['company'].drop_duplicates().sort_values(ascending=False)
df.drop_duplicates()
df.head()
```

```python
# Hypotesis 1: Budget will have high correlation with Gross revenue
# Scatterplot budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Start testing using correlation (pearson[default], kendall, and spearman)
df.corr(method='pearson')
```

```python
# High correlation Budget vs Gross - Hypotesis 1 is correct based on data
# now to visualise the correlation better
corr_matrix = df.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Hypotesis 2: Company name will have high correlation with Gross revenue
# Numerize/corelate the company name into number
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
df.head()
```

```python
# visualize the matrix to prove hypothesis 2

corr_matrix = df_numerized.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Hypothesis 2, Company have no correlation to Gross earnings
# unstacking the correlation to search for any features having high correlation
corr_matrix = df_numerized.corr(method='pearson')
corr_pairs = corr_matrix.unstack()
corr_pairs
```

```python
# sorting out the pairings
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# filter the highest positive correlation pairs
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
